```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data from the SQL export
data = pd.read_csv('sql_query_results.csv')

# Calculate statistics
metrics = {
    'Metric': ['Reg-Watch Duration', 'Watch-Purchase Duration'],
    'Mean': [data['av_reg_watch'].mean(), data['av_watch_purch'].mean()],
    'Median': [data['av_reg_watch'].median(), data['av_watch_purch'].median()],
    'Mode': [data['av_reg_watch'].mode()[0], data['av_watch_purch'].mode()[0]]
}

stats_df = pd.DataFrame(metrics)
print(stats_df)

# Distribution plots
plt.figure(figsize=(14, 6))

# Reg-Watch Distribution
plt.subplot(1, 2, 1)
sns.histplot(data['av_reg_watch'], bins=20, kde=True, color='blue')
plt.axvline(data['av_reg_watch'].mean(), color='red', linestyle='--', label='Mean')
plt.axvline(data['av_reg_watch'].median(), color='green', linestyle='-', label='Median')
plt.axvline(data['av_reg_watch'].mode()[0], color='purple', linestyle=':', label='Mode')
plt.title('Reg-Watch Duration Distribution')
plt.legend()

# Watch-Purchase Distribution
plt.subplot(1, 2, 2)
sns.histplot(data['av_reg_watch'], bins=20, kde=True, color='orange')
plt.axvline(data['av_reg_watch'].mean(), color='red', linestyle='--', label='Mean')
plt.axvline(data['av_reg_watch'].median(), color='green', linestyle='-', label='Median')
plt.axvline(data['av_reg_watch'].mode()[0], color='purple', linestyle=':', label='Mode')
plt.title('Watch-Purchase Duration Distribution')
plt.legend()

plt.tight_layout()
plt.show()
```
